Each tweet sample is composed by different columns
- id : unique tweet ID
- text : text of the tweet
- author_id : who posted the tweet
- created_at : datetime of the creation of the tweet
- lang : language of the tweet
- geo : geolocalization of the tweet
- retweet_count : number of retweets of the tweet
- reply_count : number of replies of the tweet
- like_count : number of likes of the tweet
- quote_count : number of quotes of the tweet
